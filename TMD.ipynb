## $F_{qg}^2$

```python
def F_qg_2(y2, data):
    """Computes the TMD quark-gluon correlation function F_qg^(2)."""
    print(f"----------------Computing F_qg^(2) for y2 = {y2}-----------------" )
    F1 = F(y2, data)
    F_WW1 = F_WW(y2, data)
    results = []
    
    for kt in kt_values:
        def integrand(theta, r):
            r_prime = np.sqrt(kt**2 + r**2 - 2*kt*r*np.cos(theta))
            return r*F1(r_prime)*F_WW1(r)
        
        res, error = dblquad(integrand, 0, max_kt, lambda x: 0, lambda x: 2*np.pi)
        results.append(res)
        print(f"kt = {kt} done, result = {results[-1]}")

    def F_qg_2_prime(kt):
        i = find_closest_kt_index(kt)
        return results[i]
    print(f"----------------Done computing F_qg^(2) for y2 = {y2}-----------------" )
    return F_qg_2_prime

F_qg_2_func = F_qg_2(y2, data)
```

## $F_{gg}^1$

```python
def F_gg_1(y2, data):
    """Computes the TMD gluon-gluon correlation function F_gg^(1)."""
    print(f"----------------Computing F_gg^(1) for y2 = {y2}-----------------" )
    F1 = F(y2, data)
    F_fund1 = F_fund(y2, data)
    results = []
    
    for kt in kt_values:
        def integrand(theta, r):
            r_prime = np.sqrt(kt**2 + r**2 - 2*kt*r*np.cos(theta))
            return r*F1(r_prime)*F_fund1(r)
        
        res, error = dblquad(integrand, 0, max_kt, lambda x: 0, lambda x: 2*np.pi)
        results.append(res)
        print(f"kt = {kt} done, result = {results[-1]}")

    def F_gg_1_prime(kt):
        i = find_closest_kt_index(kt)
        return results[i]
    print(f"----------------Done computing F_gg^(1) for y2 = {y2}-----------------" )
    return F_gg_1_prime

F_gg_1_func = F_gg_1(y2, data)
```

## $F_{gg}^2$

```python
def F_gg_2(y2, data):
    FF = F_gg_1_func
    def F_prime(kt):
        return FF(kt)-F_adj(y2, data)(kt)
    return F_prime

F_gg_2_func = F_gg_2(y2, data)
```

## $F * F$

```python
#intermediate function for F_gg_6
def F_F(y2, data):
    """Computes the convolution of F and F"""
    print(f"----------------Computing F_F for y2 = {y2}-----------------" )
    F1 = F(y2, data)
    F_fund1 = F1
    results = []
    
    for kt in kt_values:
        def integrand(theta, r):
            r_prime = np.sqrt(kt**2 + r**2 - 2*kt*r*np.cos(theta))
            return r*F1(r_prime)*F_fund1(r)
        
        res, error = dblquad(integrand, 0, max_kt, lambda x: 0, lambda x: 2*np.pi)
        results.append(res)
        print(f"kt = {kt} done, result = {results[-1]}")

    def F_F_prime(kt):
        i = find_closest_kt_index(kt)
        return results[i]
    print(f"----------------Done computing F_F for y2 = {y2}-----------------" )
    return F_F_prime

F_F_func = F_F(y2, data)
```

## $F_{gg}^6$

```python
def F_gg_6(y2, data):
    """Computes the TMD gluon-gluon correlation function F_gg^(6)."""
    print(f"----------------Computing F_gg^(6) for y2 = {y2}-----------------")
    F1 = F_F_func
    Fww = F_WW(y2, data)
    results = []
    
    for kt in kt_values:
        def integrand(theta, r):
            #YOU NEED TO INTROUCE THE DEPENDANCE ON Q_T' HERE
            r_prime = np.sqrt(kt**2 + r**2 - 2*kt*r*np.cos(theta))
            return r*Fww(r)*F1(r_prime)
        
        res, error = dblquad(integrand, 0, max_kt, lambda x: 0, lambda x: 2*np.pi)

        results.append(res)
        print(f"kt = {kt} done, result = {results[-1]}")

    def F_gg_6_prime(kt):
        i = find_closest_kt_index(kt)
        return results[i]
    print(f"----------------Done computing F_gg^(6) for y2 = {y2}-----------------" )
    return F_gg_6_prime

F_gg_6_func = F_gg_6(y2, data)
```

# Hard Factors

```python
# Define Python functions for each case based on the symbolic expressions

def K_gg_1(ubar, sbar, tbar, uhat, shat, that):
    return 2 * (sbar**4 + tbar**4 + ubar**4) * (ubar * uhat + tbar * that)/ (tbar * that * ubar * uhat * shat * sbar)

def K_gg_2(ubar, sbar, tbar, uhat, shat, that):
    return -1 * (sbar**4 + tbar**4 + ubar**4) * (ubar * uhat + tbar * that - shat * sbar)/ (tbar * that * ubar * uhat * shat * sbar)

def K_qq_1(ubar, sbar, tbar, uhat, shat, that):
    return (tbar**2 + ubar**2) * (ubar * uhat + tbar * that)/ (2 * N_c * sbar * shat * that * uhat)

def K_qq_2(ubar, sbar, tbar, uhat, shat, that):
    return (tbar**2 + ubar**2) * (ubar * uhat + tbar * that - shat * sbar)/ (2 * N_c**3 * sbar * shat * that * uhat)

def K_qg_1(ubar, sbar, tbar, uhat, shat, that):
    return - (ubar * (sbar**2 + ubar**2)) / (2 * tbar * that * shat)

def K_qg_2(ubar, sbar, tbar, uhat, shat, that):
    return - (sbar * (sbar**2 + ubar**2)) / (2 * tbar * that * ubar)

def hard_factors(a, c, d, i):
    if i==1:
        if a == 'q' and c == 'q' and d == 'g':
            return K_qg_1
        elif a == 'g' and c == 'q' and d == 'q':
            return K_qq_1
        elif a == 'g' and c == 'g' and d == 'g':
            return K_gg_1
    elif i==2:
        if a == 'q' and c == 'q' and d == 'g':
            return K_qg_2
        elif a == 'g' and c == 'q' and d == 'q':
            return K_qq_2
        elif a == 'g' and c == 'g' and d == 'g':
            return K_gg_2

def TMDs(a,c,d,i):
    if i==1:
        if a == 'q' and c == 'q' and d == 'g':
            return F_qg_1_func
        elif a == 'g' and c == 'q' and d == 'q':
            return F_gg_1_func
        elif a == 'g' and c == 'g' and d == 'g':
            f = lambda kt: 0.5*(F_gg_1_func(kt)+F_gg_6_func(kt))
            return f
    elif i==2:
        if a == 'q' and c == 'q' and d == 'g':
            return F_qg_2_func
        elif a == 'g' and c == 'q' and d == 'q':
            f = lambda kt: -N_c**2*F_gg_2_func(kt)
            return f
        elif a == 'g' and c == 'g' and d == 'g':
            f = lambda kt: F_gg_2_func(kt)+F_gg_6_func(kt)
            return f


```

```python
y1, y2 = 4, 4
pt1, pt2 = 20, 20
p1_plus, p1_minus = pt1*np.exp(y1)/2, pt1*np.exp(-y1)/2
p2_plus, p2_minus = pt2*np.exp(y2)/2, pt2*np.exp(-y2)/2
z = p1_plus/(p1_plus+p2_plus)

def kt_square(phi):
    return pt1**2 + pt2**2 + 2*pt1*pt2*np.cos(phi)

def Pt_sqaure(phi):
    return (1-z)**2*pt1**2 + z**2*pt2**2 - 2*z*(1-z)*pt1*pt2*np.cos(phi)

def shat(phi):
    return Pt_sqaure(phi)/(z*(1-z))

def that(phi):
    return -pt2**2/(1-z)

def uhat(phi):
    return -pt1**2/z

def sbar(phi):
    return shat(phi)+kt_square(phi)

def tbar(phi):
    return -z*sbar(phi)

def ubar(phi):
    return -(1-z)*sbar(phi)

```

```python
def nuclear_modification_factor(phi):
    uhat_value = uhat(phi)
    that_value = that(phi)
    shat_value = shat(phi)
    sbar_value = sbar(phi)
    tbar_value = tbar(phi)
    ubar_value = ubar(phi)
    kt = np.sqrt(kt_square(phi))
    
    configurations = [('q', 'q', 'g'), ('g', 'q', 'q'), ('g', 'g', 'g')]

    numerator = 0
    for config in configurations:
        a,c,d = config
        summ = hard_factors(a,c,d,1)(ubar_value, sbar_value, tbar_value, uhat_value, shat_value, that_value)*TMDs(a,c,d,1)(kt) + hard_factors(a,c,d,2)(ubar_value, sbar_value, tbar_value, uhat_value, shat_value, that_value)*TMDs(a,c,d,2)(kt)
        numerator += summ/(1+(c==d))
    
    return numerator

        
    
```
